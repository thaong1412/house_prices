# Import libraries

```python
import pandas as pd 

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns
```

# Loading data

```python
train = pd.read_csv("train.csv")
```

```python
train.head()
```

```python
train.info()
```

### Distribution of sale price

```python
plt.figure(figsize=(10, 6))  

sns.histplot(data = train, x='SalePrice', kde=True, color='skyblue', bins=50)

plt.xlabel("SalePrice")
plt.ylabel("Frequency")
plt.title("Distribution of sale price")

plt.savefig("saleprice.jpg", format="jpeg")

plt.show()
```

### Average house price per year

```python
# Change the data type of column "YrSold" from int64 to str

train['YrSold'] = train['YrSold'].astype(str)
```

```python
price_year = pd.DataFrame(train.groupby("YrSold")["SalePrice"].mean())
price_year.reset_index(inplace = True)

price_year
```

```python
# Visualization of average price per year

plt.bar(price_year["YrSold"],price_year["SalePrice"])

plt.xlabel("Avergae price")
plt.ylabel("Dollars")
plt.title("Avergae house price per year")

plt.savefig("Avergae price.jpg", format="jpeg")

plt.show()
```

### Average price of different MSSubClass group per year

```python
grouping_criteria = {
    20: '1 - story',
    30: '1 - story',
    40: '1 - story',
    45: '1 - story',
    50: '1 - story',
    60: '2 - story',
    70: '2 - story',
    75: '2 - story',
    80: 'Split',
    85: 'Split',
    90: 'Duplex/ Family conversion',
    120: '1 & 2 story Pud',
    150: '1 & 2 story Pud',
    160: '1 & 2 story Pud',
    180: '1 & 2 story Pud',
    190: 'Duplex/ Family conversion'
}

# Group the data based on the criteria
train['MSSubClassGroup'] = train['MSSubClass'].map(grouping_criteria)
```

```python
price_class_year = train.groupby(["YrSold","MSSubClassGroup"])["SalePrice"].mean().reset_index()

price_class_year
```

```python
sns.lineplot( data = price_class_year, x = "YrSold", y = "SalePrice", hue = "MSSubClassGroup", palette = "Set2")

plt.xlabel("Avergae price")
plt.ylabel("Dollars")
plt.title("Avergae house price per year")

plt.savefig("line chart.jpg", format="jpeg")

plt.show()
```

# Calculate the correlation matrix

```python
# Select only numeric columns
numeric_columns = train.select_dtypes(include=['int64', 'float64'])

correlation_matrix = numeric_columns.corr()

# Filter and display values greater than 0.5
filtered_correlation_matrix = correlation_matrix[(correlation_matrix != 1) & (correlation_matrix > 0.7)]

filtered_correlation_matrix
```

```python
# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(filtered_correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')

plt.savefig("heatmap.jpg", format="jpeg")

plt.show()
```
